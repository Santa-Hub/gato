```python
import matplotlib.pyplot as plt

# Datos simulados para los gráficos
sprints = [1, 2, 3, 4, 5, 6]
success_rate = [0.6, 0.7, 0.8, 0.85, 0.9, 0.95]  # Tasa de éxito de envió
validation_errors = [10, 8, 6, 4, 3, 1]  # Número de errores de validación
layout_verification = [0.7, 0.75, 0.8, 0.85, 0.9, 0.95]  # Verificación de Layout y Diseño
functionality_verification = [0.6, 0.7, 0.8, 0.85, 0.9, 0.95]  # Verificación de Funcionalidad
compatibility_verification = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]  # Verificación de compatibilidad
documentation_results = [0.6, 0.65, 0.7, 0.75, 0.8, 0.9]  # Documentación de Resultados
user_satisfaction = [6, 7, 8, 8.5, 9, 9.5]  # Satisfacción del usuario (1 a 10)

# Configuración de la figura
plt.figure(figsize=(15, 10))

# Tasa de éxito de envío
plt.subplot(3, 2, 1)
plt.plot(sprints, success_rate, marker='o')
plt.title("Tasa de Éxito de Envío")
plt.xlabel("Sprint")
plt.ylabel("Tasa de éxito")
plt.ylim(0, 1)

# Número de errores de validación
plt.subplot(3, 2, 2)
plt.plot(sprints, validation_errors, marker='o', color='r')
plt.title("Número de Errores de Validación")
plt.xlabel("Sprint")
plt.ylabel("Errores")
plt.ylim(0, max(validation_errors) + 2)

# Verificación de Layout y Diseño
plt.subplot(3, 2, 3)
plt.plot(sprints, layout_verification, marker='o', color='g')
plt.title("Verificación de Layout y Diseño")
plt.xlabel("Sprint")
plt.ylabel("Verificación")
plt.ylim(0, 1)

# Verificación de Funcionalidad
plt.subplot(3, 2, 4)
plt.plot(sprints, functionality_verification, marker='o', color='m')
plt.title("Verificación de Funcionalidad")
plt.xlabel("Sprint")
plt.ylabel("Verificación")
plt.ylim(0, 1)

# Verificación de compatibilidad
plt.subplot(3, 2, 5)
plt.plot(sprints, compatibility_verification, marker='o', color='c')
plt.title("Verificación de Compatibilidad")
plt.xlabel("Sprint")
plt.ylabel("Verificación")
plt.ylim(0, 1)

# Documentación de Resultados
plt.subplot(3, 2, 6)
plt.plot(sprints, documentation_results, marker='o', color='orange')
plt.title("Documentación de Resultados")
plt.xlabel("Sprint")
plt.ylabel("Documentación")
plt.ylim(0, 1)

# Satisfacción del usuario
plt.figure(figsize=(7, 5))
plt.plot(sprints, user_satisfaction, marker='o', color='purple')
plt.title("Satisfacción del Usuario")
plt.xlabel("Sprint")
plt.ylabel("Satisfacción (1-10)")
plt.ylim(0, 10)

plt.tight_layout()
plt.show()
```
