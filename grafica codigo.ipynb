```python
import matplotlib.pyplot as plt

# Datos simulados para los grficos
sprints = [1, 2, 3, 4, 5, 6]
success_rate = [0.6, 0.7, 0.8, 0.85, 0.9, 0.95]  # Tasa de xito de envi
validation_errors = [10, 8, 6, 4, 3, 1]  # Nmero de errores de validacin
layout_verification = [0.7, 0.75, 0.8, 0.85, 0.9, 0.95]  # Verificacin de Layout y Diseo
functionality_verification = [0.6, 0.7, 0.8, 0.85, 0.9, 0.95]  # Verificacin de Funcionalidad
compatibility_verification = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]  # Verificacin de compatibilidad
documentation_results = [0.6, 0.65, 0.7, 0.75, 0.8, 0.9]  # Documentacin de Resultados
user_satisfaction = [6, 7, 8, 8.5, 9, 9.5]  # Satisfaccin del usuario (1 a 10)

# Configuracin de la figura
plt.figure(figsize=(15, 10))

# Tasa de xito de envo
plt.subplot(3, 2, 1)
plt.plot(sprints, success_rate, marker='o')
plt.title("Tasa de xito de Envo")
plt.xlabel("Sprint")
plt.ylabel("Tasa de xito")
plt.ylim(0, 1)

# Nmero de errores de validacin
plt.subplot(3, 2, 2)
plt.plot(sprints, validation_errors, marker='o', color='r')
plt.title("Nmero de Errores de Validacin")
plt.xlabel("Sprint")
plt.ylabel("Errores")
plt.ylim(0, max(validation_errors) + 2)

# Verificacin de Layout y Diseo
plt.subplot(3, 2, 3)
plt.plot(sprints, layout_verification, marker='o', color='g')
plt.title("Verificacin de Layout y Diseo")
plt.xlabel("Sprint")
plt.ylabel("Verificacin")
plt.ylim(0, 1)

# Verificacin de Funcionalidad
plt.subplot(3, 2, 4)
plt.plot(sprints, functionality_verification, marker='o', color='m')
plt.title("Verificacin de Funcionalidad")
plt.xlabel("Sprint")
plt.ylabel("Verificacin")
plt.ylim(0, 1)

# Verificacin de compatibilidad
plt.subplot(3, 2, 5)
plt.plot(sprints, compatibility_verification, marker='o', color='c')
plt.title("Verificacin de Compatibilidad")
plt.xlabel("Sprint")
plt.ylabel("Verificacin")
plt.ylim(0, 1)

# Documentacin de Resultados
plt.subplot(3, 2, 6)
plt.plot(sprints, documentation_results, marker='o', color='orange')
plt.title("Documentacin de Resultados")
plt.xlabel("Sprint")
plt.ylabel("Documentacin")
plt.ylim(0, 1)

# Satisfaccin del usuario
plt.figure(figsize=(7, 5))
plt.plot(sprints, user_satisfaction, marker='o', color='purple')
plt.title("Satisfaccin del Usuario")
plt.xlabel("Sprint")
plt.ylabel("Satisfaccin (1-10)")
plt.ylim(0, 10)

plt.tight_layout()
plt.show()
```
